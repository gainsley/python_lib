```python
class Graph(object):
    def __init__( self, _dict=None ):
        if _dict is None:
            _dict={}     
        self.g = _dict
    
    def nodes( self ):
        return self.g.keys() # returns a list
    
    def neighbourgs(self, node):
        if node in self.g.keys():
            return self.g[node]
        else:
            return None
    
    def edges( self ):
        e=[]
        for node in self.g:
            l = [ node ] * len( self.g[node] )
            tmp = zip( l, self.g[node] )
            e += tmp 
            #print "debug",l,node,len( self.g[node]), self.g[node]
        return e
    
    def add_node( self, node, neighbourg=None ):
        if neighbourg is None:
            neighbourg=[]
        if not node in self.g.keys():
            self.g[node] = neighbourg
        else:
            print "error: node already in",node
    
    def remove_node( self, node ):
        if not node in self.g.keys():
            return
        # remove node from all lists
        for n in self.g.keys():
            self.remove_edge( n, node )
        # remove key from dict
        del self.g[node]  
        
    def add_edge( self, a, b, cost=1 ):
        if a in self.g.keys() and b in self.g.keys():
            self.g[a].append ( b )
        else:
            print "error: nodes don't exist"
        
    def remove_edge( self, node, child ):
        if not node in self.g.keys():
            print "error: nodes don't exist"
            return
        if child in self.g[node]:
            self.g[node].remove(child)
            
    def __str__( self ):  
        return str( self.g.keys() )
   
    
def find_path( graph, start, end,  path=[] ):
    path = path + [start]
    if start == end:
        return path
    if not start in graph.nodes():
        return None
    for node in graph.neighbourgs(start):
        if node not in path:
            newpath = find_path( graph, node, end, path )
            if newpath: return newpath
    return None   
 
    
def find_all_path( graph, start, end,  path=[] ):
    path = path + [start]
    if start == end:
        return [path]
    if not start in graph.nodes():
        return []
    paths = []
    for node in graph.neighbourgs(start):
        if node not in path:
            newpaths = find_all_path( graph, node, end, path )
            for newpath in newpaths:
                paths.append(newpath)
    return paths   

        
def find_shortest_path( graph, start, end, path=[] ):
    path = path + [start]
    if start == end:
        return path
    if not start in graph.nodes():
        return None
    shortest = None
    for node in graph.neighbourgs(start):
        if node not in path:
            newpath = find_shortest_path(graph, node, end, path)
            if newpath:
                if not shortest or len(newpath) < len(shortest):
                    shortest = newpath
    return shortest


def mst( graph ):
    L = []
    graph.nodes
    
def graph_search( graph ):
    
    return
    
##---TODO-
def isolated( graph ) :
    L = []
    return L

def generate_graph( edges ) :
    graph = Graph()
    for a,b in edges:
        #print "generate_full_connected_graph", a,b
        #graph.add_node(a)
        #graph.add_node(b)
        #graph.add_edge(a,b)
        graph.add_node(b)
        graph.add_node(a, [b])
    return graph
```
